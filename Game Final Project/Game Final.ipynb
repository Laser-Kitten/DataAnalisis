## Проект: Анализ рекламных источников

С учётом планирующейся монетизации игры (показ рекламы на экране с выбором типа объекта для постройки) и исходя из названия проекта, предполагаем, что данное исследование проводится в интересах отдела маркетинга. Планируемый результат работы - приоритезация каналов привлечения клиентов, оптимизация маркетинговых расходов в разрезе будущей монетизации.

Задача: На основании поведения юзеров оценить маркетинговую привлекательность каналов привлечения пользователей. Оценка будет основываться на количестве построенных юзером объектов в разрезе источника, так как чем больше объектов игрок построил, тем больше рекламного трафика было им потреблено при выборе объекта. Таким образом наша метрика - соотношение затрат на привлечение пользователя с количеством построенных объектов. 

На основании имеющейся задачи дополним наше ТЗ дополнительной гипотезой **о равенстве количества построенных объектов в зависимости от источника**.

## Предобработаем данные

#### Обработка названий значений 

```python
#заменим названия зданий
df_actions['building_type'] = df_actions['building_type'].replace({'assembly_shop':'цех',
                                             'spaceport':'космопорт',
                                             'research_center':'лаборатория',
                                             })
#заменим названия источников трафика
df_costs['source'] = df_costs['source'].replace({'facebook_ads':'facebook',
                                             'instagram_new_adverts':'instagram',
                                             'yandex_direct':'yandex',
                                             'youtube_channel_reklama':'youtube',  
                                             })

df_source['source'] = df_source['source'].replace({'facebook_ads':'facebook',
                                             'instagram_new_adverts':'instagram',
                                             'yandex_direct':'yandex',
                                             'youtube_channel_reklama':'youtube',  
                                             })
```

#### Преобразование типов данных

Понизим битность у цифровых данных и заменим тип данных столбцов с датами.

```python
#используем понижающее приведение downcast метода pd.to_numeric() 
df_costs['cost'] = pd.to_numeric(df_costs['cost'], downcast='float')

df_actions['event_datetime'] = pd.to_datetime(df_actions['event_datetime'], format='%Y-%m-%d %H:%M:%S')

#добавим столбцы с датами
df_actions['event_date'] = df_actions['event_datetime'].dt.date
df_actions['event_week']  = df_actions['event_datetime'].dt.week
df_actions['year']  = df_actions['event_datetime'].dt.year

#проверим как изменились типы данных методом info()
df_costs.info()
df_actions.info()
```

#### Обработка дубликатов

```python
#суммируем количество дубликатов
duplicated = df_actions.duplicated().sum()
print('Количество полных дубликатов составляет\033[1m',duplicated,'\033[0m')
print('Доля полных дубликатов составляет\033[1m {:.2%}\033[0m'.format(duplicated / len(df_actions)))
```

```python
#суммируем количество дубликатов
duplicated = df_costs.duplicated().sum()
print('Количество полных дубликатов составляет\033[1m',duplicated,'\033[0m')
print('Доля полных дубликатов составляет\033[1m {:.2%}\033[0m'.format(duplicated / len(df_costs)))
```

```python
#суммируем количество дубликатов
duplicated = df_source.duplicated().sum()
print('Количество полных дубликатов составляет\033[1m',duplicated,'\033[0m')
print('Доля полных дубликатов составляет\033[1m {:.2%}\033[0m'.format(duplicated / len(df_costs)))
```

```python
#удалим дубликаты методом drop_duplecates() 
df_actions = df_actions.drop_duplicates()
duplicated = df_actions.duplicated().sum()
print('Количество полных дубликатов составляет\033[1m', duplicated, '\033[0m')
```

#### Исследуем пропущенные значения

Посмотрим имеются ли закономерности в пропущенных значениях в столбце с типом построенного здания.

```python
# срез df с условием пустых значений в столбце building_type
null_build = df_actions[df_actions['building_type'].isnull()]
display(null_build.head())
# проверим какие события генерируют NaN
null_build['event'].value_counts()
```

Явно, что пропуски в столбце с типом построенного здания не являются ошибкой данных, они связаны с событиями завершения проекта или первого уровня. Теперь изучим пустые значения в столбце `project_type`.

```python
# срез df с условием отсутствия пустых значений в столбце project_type
project = df_actions.dropna(subset=['project_type'])   
display(project.head())
# проверим какие события генерируют NaN
project['event'].value_counts()
```

Таким образом наличие значения в столбце `project_type` связано с событием завершения проекта в столбце `event`, в остальных случаях содержится NaN. В остальных имеющихся таблицах отсутствуют пропуски.

## Исследовательский анализ данных (EDA)

```python
print('')
print('Общее количество событий равно\033[1m', len(df_actions['event']),'\033[0m')

print('')
print('Количество уникальных пользователей составляет\033[1m', df_actions['user_id'].nunique(),'\033[0m')

print('')
print('В среднем на одного пользователя приходится \033[1m{:.0f}\033[0m событий'
      .format(df_actions.groupby('user_id')['event'].agg('count').median()))
```

#### Оценим возможные выбросы в количестве событий по пользователям

```python
event_per_user = df_actions.groupby('user_id')['event'].count().reset_index()

#получим данные о распределении группировки методом describe()
event_per_user['event'].describe()
```

```python
# визуализируем 
plt.figure(figsize=(16, 7))
sns.set_palette('coolwarm')
x_values = pd.Series(range(0,len(event_per_user['event'])))
plt.scatter(x_values, event_per_user['event']);
plt.title('График распределения количества событий по пользователям')
plt.ylabel('Количество событий')
plt.show()

plt.figure(figsize=(16, 7))
sns.set_palette('bright')
sns.distplot(event_per_user['event'], bins=70) 
plt.title('Распределение количества событий по пользователям')
plt.xlabel('Среднее количество событий')
plt.ylabel('')
plt.show()
```

```python
#рассчитаем 99 персентиль для определения границы выбросов 
print('На 99% пользователей приходится менее \033[1m{:.0f}\033[0m событий'
      .format(event_per_user['event'].quantile(0.99)))

unnormal_id = list(event_per_user.query('event > 19')['user_id'])
print('Количество юзеров с более чем 19 событиями равно\033[1m', len(unnormal_id),'\033[0m')
```

Как мы можем заметить среднее **медианное значение** событий на одного пользователя равно **10**. **Минимальное равно 1** событию, а **максимальное 22**. **На 99 процентов** пользователей приходится **менее 19** событий (99 квантиль нашего распределения). #### Рассчитаем Retention rate

Важно оценить коэффициэнт удержания, насколько игроки увлекаются игрой и как долго продолжают в неё играть и не уходят. Исключим из выборки игроков, которые перешли на второй уровень, потому что они в процессе завершения уровня будут исчезать из датафрейма, так как не будет сведений о событиях.

```python
# объедим таблицы
df_actions = df_actions.merge(df_source, on = 'user_id')
```

```python
#получим для каждого игрока время его первого события
first_event_datetime = df_actions.groupby('user_id').agg({'event_datetime': 'min'}).reset_index()
first_event_datetime.columns = ['user_id', 'first_event_datetime']
# добавим столбец к df
df_actions = df_actions.merge(first_event_datetime, on = 'user_id')
```

```python
visits = df_actions
# добавим столбцы с первым событием игрока и день каждого события
visits['first_visit_day'] = visits['first_event_datetime'].astype('datetime64[D]')
visits['visit_day'] = visits['event_datetime'].astype('datetime64[D]')
```

```python
# создадим список с юзерами, завершившими уровень
user_level_done = list(df_actions[df_actions['event'] == 'finished_stage_1']['user_id'])
#исключим юзеров завершивших уровень из df
visits = visits[~visits.user_id.isin(user_level_done)] 
```

```python
#отнимем месяц визита, от первого месяца посещения, затем разделим на недельную константу
visits['lifetime'] = visits['visit_day'] - visits['first_visit_day'] 
visits['lifetime'] = visits['lifetime'] / np.timedelta64(1, 'W')
visits['lifetime'] = visits['lifetime'].round().astype('int')

#сгруппируем уникальных пользователей по месяцу первого визита и времени жизни когорты
cohorts = visits.groupby(['first_visit_day','lifetime']).agg({'user_id':'nunique'}).reset_index() 

#найдём исходное количество пользователей в когорте (на нулевой месяц)
initial_users_count = cohorts[cohorts['lifetime'] == 0][['first_visit_day','user_id']]
initial_users_count = initial_users_count.rename(columns={'user_id':'cohort_uid'})

#объединим с исходной группировкой
cohorts = cohorts.merge(initial_users_count, on = 'first_visit_day')
#рассчитаем Retention
cohorts['retention'] = cohorts['user_id'] / cohorts['cohort_uid']

#построим сводную таблицу 
retention_pivot = cohorts.pivot_table(index='first_visit_day', columns = 'lifetime', 
                                      values='retention',aggfunc='sum')

#посмотрим на сводную таблицу с помощью .heatmap()
plt.figure(figsize=(15, 6))
plt.title('Retention игроков по недельным событиям в игре')
ax = sns.heatmap(retention_pivot, annot=True, fmt='.2%', linewidths=1, linecolor='gray')
ticklabels = [retention_pivot.index[int(tick)].strftime('%Y-%m-%d') for tick in ax.get_yticks()]
ax.set_yticklabels(ticklabels)
plt.xlabel('week lifetime')
plt.show()
```

```python
second_week_lifetime = retention_pivot.reset_index(drop=True)[1]
print('Cредний Retention Rate на вторую неделю «жизни» пользователей \033[1m{:.2%}\033[0m'
      .format(second_week_lifetime.mean()))
third_week_lifetime = retention_pivot.reset_index(drop=True)[2]
print('Cредний Retention Rate на третью неделю «жизни» пользователей \033[1m{:.2%}\033[0m'
      .format(third_week_lifetime.mean()))
```

Пользователи, которые не завершили первый уровень начинают уходить на третьей неделе игры, на четвертой практически их не остается. Причины мы выявить не можем, может сложность становится слишком высокой или уровень слишком долгий, но эту информацию можно использовать для каких-то продуктовых решений и удержания пользователей в игре.

Как мы видим, наиболее успешные проведенные маркетинговые кампании по Ютуб и ЯндексДирект.

```python
# создадим список с источниками 
sources = list(df_actions['source'].unique())

for source in sources:
    visits_by_event = visits[visits['source'] == source]
        
    #сгруппируем уникальных пользователей по месяцу первого визита и времени жизни когорты
    cohorts = visits_by_event.groupby(['first_visit_day','lifetime']).agg({'user_id':'nunique'}).reset_index() 

    #найдём исходное количество пользователей в когорте (на нулевой месяц)
    initial_users_count = cohorts[cohorts['lifetime'] == 0][['first_visit_day','user_id']]
    initial_users_count = initial_users_count.rename(columns={'user_id':'cohort_uid'})

    #объединим с исходной группировкой
    cohorts = cohorts.merge(initial_users_count, on = 'first_visit_day')
    #рассчитаем Retention
    cohorts['retention'] = cohorts['user_id'] / cohorts['cohort_uid']

    #построим сводную таблицу 
    retention_pivot = cohorts.pivot_table(index='first_visit_day', columns = 'lifetime', 
                                          values='retention',aggfunc='sum')

    #посмотрим на сводную таблицу с помощью .heatmap()
    plt.figure(figsize=(15, 5))
    plt.title('Retention игроков источника ' + source)
    ax = sns.heatmap(retention_pivot, annot=True, fmt='.2%', linewidths=1, linecolor='gray')
    ticklabels = [retention_pivot.index[int(tick)].strftime('%Y-%m-%d') for tick in ax.get_yticks()]
    ax.set_yticklabels(ticklabels)
    plt.xlabel('week lifetime')
    plt.show()
    
    second_week_lifetime = retention_pivot.reset_index(drop=True)[1]
    third_week_lifetime = retention_pivot.reset_index(drop=True)[2]
    print('Cредний Retention Rate на вторую неделю «жизни» пользователей источника {} \033[1m{:.2%}\033[0m'
      .format(source, second_week_lifetime.mean()))
    print('Cредний Retention Rate на третью неделю «жизни» пользователей источника {} \033[1m{:.2%}\033[0m'
      .format(source, third_week_lifetime.mean()))
 
```

Значения коэффициента удержания среди пользователей, на прошедших на второй уровень, в разрезе источников привлечения практически не отличается, хотя есть небольшое лидерство за fb и instagram.

### Одинаково ли время завершения уровня в зависимости способа прохождения

Для ответа на первый вопрос нам нужно проверить гипотезу о равенстве среднего двух генеральных совокупностей (время завершения уровня в зависимости способа прохождения (победа над игроком или завершение проекта) всеми игроками нашей игры) по взятым из них выборкам (имеющиеся у нас данные по когорте пользователей). Начнём с формулировки нулевой гипотезы **H₀ - «средние значения времени завершения уровня в зависимости от способа прохождения равно»**, соответственно альтернативная гипотеза **H₁ - «средние значения времени завершения уровня в зависимости от способа прохождения различается»**.


```python
# оставим записи только о юзерах, завершивших уровень
df_level_done = df_actions[df_actions['event'] == 'finished_stage_1']
# добавим столбец с временем от первого события игрока до завершения уровня 1
df_level_done['leveldone_time'] = df_level_done['event_datetime'] - df_level_done['first_event_datetime']
# добавим столбец со значением завершения уровня в часах
df_level_done['leveldone_hours'] = df_level_done['leveldone_time'] / np.timedelta64(1, 'h')
```

```python
# выделим df с юзерами, завершивших проект
project_win = df_actions[df_actions['project_type'] == 'satellite_orbital_assembly']
user_project_win = project_win[['user_id','project_type']]
user_project_win.columns = ['user_id', 'win_type']
```

```python
# добавим столбец к df
df_level_done = df_level_done.merge(user_project_win, on = 'user_id', how='left')
```

```python
# для пользователей, завершивших уровень победой добавим соответствующее значение
df_level_done['win_type'] = df_level_done['win_type'].fillna('victory_over_enemy')
```

```python
# разделим df-мы по типу победы
u_project_win = df_level_done[df_level_done['win_type'] == 'satellite_orbital_assembly']
u_enemy_win = df_level_done[df_level_done['win_type'] == 'victory_over_enemy']
```

```python
print('')
print ('\033[1m' + 'Вывод метода describe() времени с завершением уровня проектом' + '\033[0m')
display(pd.DataFrame(u_project_win['leveldone_time'].describe()))
print('-----------------------------------------------------------------------------------------')

print('')
print ('\033[1m' + 'Вывод метода describe() времени с завершением уровня победой над соперником' + '\033[0m')
display(pd.DataFrame(u_enemy_win['leveldone_time'].describe()))
```

Игроки, завершившие 1 уровень реализацией проекта орбитальной станции, в среднем потратили это на 3 дня больше, чем игроки одержавшие победу над другими игроками - 13 дней против 10 дней, соответственно. Необходимо оценить статистическую значимость данного отличия.

```python
#визуализируем распределение выборки для оценки нормальности
plt.figure(figsize=(15, 6))
sns.distplot(u_enemy_win['leveldone_hours'], bins=35)
sns.distplot(u_project_win['leveldone_hours'], bins=35)
plt.title('Распределение времени завершения уровня разными способами')
plt.xlabel('Часы')
plt.ylabel('')
plt.legend(labels=['победой над врагом','успешным проектом'])
plt.show()
```

```python
#вычислим дисперсию выборок для корректного определения атрибута equal_val в Т–тесте
print(u_project_win['leveldone_hours'].var(ddof=1))
print(u_enemy_win['leveldone_hours'].var(ddof=1))
```

Так как мы имеем ограниченный набор статистических данных для анализа, то распределение соответствующее нашим гипотезам будет являться распределением Стьюдента (t-распределением). Поэтому в качестве статистического теста для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам будем использовать t-тест - метод scipy.stats.ttest_ind() для несвязанных выборок. Условия для Т-теста соблюдены, распределение выборок приближено к нормальному. Проверка дисперсий указывает нам, что в Т-тесте атрибут equal_val определим как False. 

```python
alpha = .05 # критический уровень статистической значимости
            # если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(u_project_win['leveldone_hours'],u_enemy_win['leveldone_hours'],equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("\033[1mОтвергаем нулевую гипотезу\033[0m, следовательно средние \
значения времени прохождения в зависимости от способа завершения различаются")
else:
    print("\033[1mНе получилось отвергнуть нулевую гипотезу\033[0m, следовательно средние \
значения времени прохождения вне зависимости от способа завершения равны") 
```

Таким образом, учитывая выявленную закономерность, что завершение уровня проектом как занимает как больше времени (13 дней против 10 дней), так и является менее популярным (1866 игроков против 3951), возможно будет правильным решением принять продуктовые решения относительно данного способа завершения уровня и сделать его более привлекательным для игроков.

### Одинаково ли количество построенных объектов пользователями в зависимости от источника

Из этого вопроса будет исходить формулировка нашей гипотезы, которую мы сформулируем в интересах отдела маркетинга. Как отмечалось ранее, планируется монетизация игры. Мы уже оценили маркетинговую привлекательность каналов привлечения пользователей посредством метрики CAC, также рассмотрели количество пользователей привлеченных с разных каналов. 

Монетизация будет осуществляться последством показа рекламы в окне выбора объекта постройки. Таким образом, чем больше объектов игрок построил, тем больше рекламного трафика было им потреблено при выборе объекта. Поэтому, метрика - соотношение затрат на привлечение пользователя с конкретного источника с количеством построенных объектов. Значит, нам нужно оценить средние значения количества построенных объектов в разрезе источников, так как большее количество зданий зданий принесут больше прибыли.


Как мы установили ранее, в разрезе источников, количество привлеченных пользователей и количество построек различны между собой. Поэтому мы будем сравнивать средние значения построек на человека. Таким образом перед нами стоит задача проверить гипотезу о равенстве среднего двух генеральных совокупностей (количество построек всеми игроками нашей игры в разрезе источников) по взятым из них выборкам (имеющиеся у нас данные по когорте пользователей в разрезе источников).

Сформулируем гипотезы:

1. Нулевая гипотеза **H₀ - «средние значения количества построек на 1 пользователя по источникам равно»**
2. Альтернативная гипотеза **H₁ - «средние значения количества построек на 1 пользователя по источникам различаются»**

```python
# посчитаем количество построек для каждого id
building_grouped = df_actions.groupby(['user_id'])[['building_type']].count().reset_index()
building_grouped.columns = ['user_id', 'building_count']
```

```python
df_actions = df_actions.merge(building_grouped, on = 'user_id', how='left')
```

```python
# разделим df-мы по типу источника
youtube_df = df_actions[df_actions['source'] == 'youtube']
facebook_df = df_actions[df_actions['source'] == 'facebook']
instagram_df = df_actions[df_actions['source'] == 'instagram']
yandex_df = df_actions[df_actions['source'] == 'yandex']
```

```python
#визуализируем распределение выборки для оценки нормальности
plt.figure(figsize=(13, 6))
sns.distplot(youtube_df['building_count'], bins=20)
sns.distplot(facebook_df['building_count'], bins=20)
plt.title('Распределение количества построек на пользователя по источникам')
plt.xlabel('Постройки')
plt.ylabel('')
plt.legend(labels=['youtube','facebook'])
plt.show()
```

```python
#визуализируем распределение выборки для оценки нормальности
plt.figure(figsize=(13, 6))
sns.distplot(instagram_df['building_count'], bins=20)
sns.distplot(yandex_df['building_count'], bins=20)
plt.title('Распределение количества построек на пользователя по источникам')
plt.xlabel('Постройки')
plt.ylabel('')
plt.legend(labels=['instagram','yandex'])
plt.show()
```

```python
#вычислим дисперсию выборок для корректного определения атрибута equal_val в Т–тесте
print(youtube_df['building_count'].var(ddof=1))
print(facebook_df['building_count'].var(ddof=1))
print(instagram_df['building_count'].var(ddof=1))
print(yandex_df['building_count'].var(ddof=1))
```

Так как мы имеем ограниченный набор статистических данных для анализа, то распределение соответствующее нашим гипотезам будет являться распределением Стьюдента (t-распределением). Поэтому в качестве статистического теста для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам будем использовать t-тест - метод scipy.stats.ttest_ind() для несвязанных выборок. Условия для Т-теста соблюдены, распределение выборок приближено к нормальному. Проверка дисперсий указывает нам (хоть значения и очень близки, но не идентичны), что в Т-тесте атрибут equal_val определим как False. 

```python
def t_stat(source1_df,source2_df):
    alpha = .05 # критический уровень статистической значимости
    bonferroni_alpha = alpha / 6 # будет произведено 6 сравнений

    results = st.ttest_ind(source1_df['building_count'],source2_df['building_count'],equal_var = False)
    
    print('p-значение:', results.pvalue)

    if results.pvalue < bonferroni_alpha:
        print("\033[1mОтвергаем нулевую гипотезу\033[0m, следовательно средние \
значения количества построек на 1 игрока в зависимости от источника различаются")
    else:
        print("\033[1mНе получилось отвергнуть нулевую гипотезу\033[0m, следовательно средние \
значения количества построек на 1 игрока в зависимости от источника равны")
        
    print('')
```

```python
t_stat(youtube_df,facebook_df)
t_stat(youtube_df,instagram_df)
t_stat(youtube_df,yandex_df)
t_stat(facebook_df,instagram_df)
t_stat(facebook_df,yandex_df)
t_stat(instagram_df,yandex_df)
```

```python
print('Среднее количество построек на пользователя источника youtube', youtube_df['building_count'].mean())
print('Среднее количество построек на пользователя источника facebook',facebook_df['building_count'].mean())
print('Среднее количество построек на пользователя источника instagram',instagram_df['building_count'].mean())
print('Среднее количество построек на пользователя источника yandex',yandex_df['building_count'].mean())
```

На основании изложенного, мы можем сделать вывод, что нет статистически значимой разницы в количестве зданий, которые строят пользователи, только между Youtube и Yandex. Значения ж по остальным источникам имеют статистическую значимость. Среднее количество построек на пользователя по  FB и Instagram, чуть выше. Но отметим, что значение совсем невелико, и это не будет влияять на грядущую монетизацию игры существенным образом. Следовательно при оценке по какому каналу привлекать пользователей скорее всего придется иными способами, например опираясь на метрику САС, которую мы рассчитали ранее.
